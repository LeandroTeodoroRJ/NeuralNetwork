```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
#Separando os atributos de entrada e a classe alvo
entradas = iris.data
entradas
```

```python
saidas = iris.target
saidas
```

```python
#Cria a rede
redeNeural = MLPClassifier(
                verbose = True,      #Mostra o erro de cada iteração :: loss = erro (loss function)
                max_iter = 1000,     #Máximo número de iterações 
                tol = 0.00001,       #Mínimo aceitável de diminuição do erro entre épocas
                activation = 'logistic',   #Função de ativação, ReLu é a default
                #hidden_layer_sizes=(5, 2) #2 camadas ocultas com 5 e 2 neurônios
                hidden_layer_sizes=(5)  #1 camada oculta com 5 neurônios
             )  
```

```python
#Faz o treinamento
redeNeural.fit(entradas, saidas)
```

Erro alto devido a estudo sobre alteração da camada escondida

```python
#Testes de predição
redeNeural.predict([[5, 7.2, 5.1, 2.2]])
```

```python
#Checando a classe predita
iris.target_names
```

```python
#Retorna alguns parâmetros da rede neural
redeNeural.get_params()
```

```python
#Estrutura da rede neural
[coef.shape for coef in redeNeural.coefs_]
```

Note:
    4 neurônios na entrada
    5 neurônios na camada oculta
    3 neurônios na camada de saída
